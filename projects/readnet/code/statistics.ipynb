```python
import pandas as pd
import re
```

```python
# CONFIG
dataset_name = "kennedy_james" # "English_ECSC"
subsection = "freespeech" #"freespeech" # "men", "women", "child"
```

```python
data_folder = "../data"
dataset_name = f"{dataset_name}_{subsection}_dataset" # https://childes.talkbank.org/access/Frogs/English-ECSC.html
output_data_directory_name = "original_audio_segments"
output_xlsx_file_name = "segments.xlsx"
```

```python
filename = f'{data_folder}/{dataset_name}/{output_data_directory_name}/{output_xlsx_file_name}'
```

```python
df = pd.read_excel(filename)
df.rename(columns={'speaker_id': 'participant'}, inplace=True)
```

```python
df
```

```python
# Function to count words in a text
def count_words(text):
    words = re.findall(r'\w+', text)
    return len(words)
```

```python
# Number of participants
num_participants = df['participant'].nunique()

# Drop duplicate rows based on 'participant' column
df_unique_participants = df.drop_duplicates(subset='participant', keep='first')

# Age statistics
try:
    mean_age = df_unique_participants['age'].mean()
    std_age = df_unique_participants['age'].std(ddof=0)
except Exception as e:
    print("HOUSTON!")
    mean_age = None
    std_age = None
    
# Number of males and females
sex_count = df_unique_participants['sex'].value_counts()

try:    
    # Number of native and non-native speakers
    native_status_count = df_unique_participants['native_status'].value_counts()
except Exception as e:
    print("HOUSTON!")
    native_status_count = None
    
# Duration statistics
tot_duration = df['duration_in_ms'].sum()
mean_duration = df['duration_in_ms'].mean()
std_duration = df['duration_in_ms'].std(ddof=0)

# Number of files
num_files = df.shape[0]

# Files per participant
files_per_participant = df.groupby('participant')['path_to_audio_segment_file'].count()
mean_files_per_participant = files_per_participant.mean()
std_files_per_participant = files_per_participant.std(ddof=0)

# Text statistics
df['word_count'] = df['text'].apply(count_words)
total_words = df['word_count'].sum()
mean_words_per_file = df['word_count'].mean()
std_words_per_file = df['word_count'].std(ddof=0)
mean_words_per_participant = df.groupby('participant')['word_count'].mean()
std_words_per_participant = df.groupby('participant')['word_count'].std(ddof=0)    
    
# Display the results
print("Number of Participants:", num_participants)
print("Mean Age:", mean_age)
print("Standard Deviation of Age:", std_age)
print("Sex Count:\n", sex_count)
print("Tot Duration (in ms):", tot_duration)
print("Mean Duration (in ms):", mean_duration)
print("Standard Deviation of Duration (in ms):", std_duration)
print("Number of Files:", num_files)
print("Mean Files per Participant:", mean_files_per_participant)
print("Standard Deviation of Files per Participant:", std_files_per_participant)
print("Total Words:", total_words)
print("Mean Words per File:", mean_words_per_file)
print("Standard Deviation of Words per File:", std_words_per_file)
print("Native Status Count:\n", native_status_count)

#print("Mean Words per Participant:\n", mean_words_per_participant)
#print("Standard Deviation of Words per Participant:\n", std_words_per_participant)
```
